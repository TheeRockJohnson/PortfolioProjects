```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from matplotlib.pyplot import figure

df = pd.read_csv(r"C:\Users\DIMA7\Downloads\archive (2)\movies.csv")
```

```python
df.dtypes
```

```python
dfe = df.drop_duplicates()
dfe
```

```python
dfe = dfe.fillna(0)
```

```python
for col in dfe.columns:
    pct_missing = np.mean(dfe[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
dfe['budget'] = dfe['budget'].astype('int64')
dfe['gross'] = dfe['gross'].astype('int64')
dfe['votes'] = dfe['votes'].astype('int64')
```

```python
dfe
```

```python
dfe['year adjusted'] = dfe['released'].astype(str)
dfe['year adjusted'] = dfe['year adjusted'].str.split(',', n=1).str[1].str.strip().fillna("")
dfe['year adjusted'] = dfe['year adjusted'].str.split('(', n=1).str[0].str.strip().fillna("")
dfe['year adjusted'] = dfe['year adjusted'].replace('',0)
dfe['year adjusted'] = dfe['year adjusted'].astype('int64')
```

```python
dfe.dtypes
```

```python
dfe['year equal'] = (dfe['year'] == dfe['year adjusted'])
```

```python
dfe
```

```python
indices_to_drop = dfe[dfe['year equal'] == False].index
dfe.drop(indices_to_drop, inplace=True)

# или просто 
# dfe = dfe[dfe['year equal']]
# dfe
```

```python
are_identical = dfe['year'].equals(dfe['year adjusted'])
are_identical
```

```python
true_count = dfe['year equal'].value_counts()[False]
true_count
```

```python
dfe.sort_values('gross', inplace = False, ascending = False)
```

```python
pd.set_option('display.max_rows',20)
```

```python
plt.style.use('ggplot')
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
plt.scatter(x = dfe['budget'], y = dfe['gross'])
plt.title('Budget vs Gross')
plt.xlabel('Gross')
plt.ylabel('Budget')

plt.show()
```

```python
sns.regplot(x = 'budget', y = 'gross', data = dfe, scatter_kws={'color': 'red'}, line_kws={'color': 'black'})
```

```python
correlation_matrix = dfe.corr(method = 'pearson', numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation_matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movies Features')
# plt.rcParams['xtick.labelsize'] = 8
# matplotlib.rcParams['figure.labelsize'] = 100

plt.show()
```

```python
dfe_num = dfe

for col in dfe_num.columns:
    if (dfe_num[col].dtype == 'object'):
        dfe_num[col] = dfe_num[col].astype('category')
        dfe_num[col] = dfe_num[col].cat.codes
dfe_num
```

```python
correlation_matrix = dfe_num.corr(method = 'pearson', numeric_only = True)

sns.heatmap(correlation_matrix, annot = True)

plt.title('Correlation_matrix')
plt.xlabel('Movie Features')
plt.ylabel('Movies Features')

plt.show()
```
